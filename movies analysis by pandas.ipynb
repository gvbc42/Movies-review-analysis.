```python
# Convert the Budject of movies in to millions
movies["budget_milions"]=movies["budget"]/10**6
movies["Gross_millions"]=movies["Gross"]/10**6
movies
```

```python
# Profit of each move title wise
movies["Profit"]=(movies["Gross_millions"]-movies["budget_milions"])/movies["budget_milions"]*100
movies
```

```python
#acotor wise face book likes
movies_3roles['Facebook_likes']=movies_3roles['actor_1_facebook_likes']+movies_3roles['actor_2_facebook_likes']+movies_3roles['actor_3_facebook_likes']
movies_3roles
```
